### What is the average(mean) number of nights for a stay across the entire data set? Fell free to round this to 2 decimal points

```python
# Add new column
hotels_df['total_stay_nights'] = hotels_df['stays_in_week_nights'] + hotels_df['stays_in_weekend_nights']
round(hotels_df['total_stay_nights'].mean(),2)
```

### What is the average total cost for a stay in the dataset? Not average dailly cost, but total stay cost. Feel free to round this to 2 decimal number.

```python
hotels_df['total_spend'] = hotels_df['adr'] * hotels_df['total_stay_nights']
round(hotels_df['total_spend'].mean(), 2)
```

### Create a table for counts for each day of the week that people arrived. (E.g: 5000 arrivals ware on a monday, 3000 arrivals ware on sunday...)

```python
import numpy as np

# Need to convert to as day, month, year
def convert(day, month, year):
    return f"{day}-{month}-{year}"
```

```python
hotels_df['date'] = np.vectorize(convert)(hotels_df['arrival_date_day_of_month'],
                                            hotels_df['arrival_date_month'], 
                                            hotels_df['arrival_date_year'])
hotels_df['date'] = pd.to_datetime(hotels_df['date'])
hotels_df['date'].dt.day_name().value_counts()
```

```python
total_amt_paid = df['tip'] + df['total_bill']
total_amt_paid
```

```python
# add the newly created column named "total_amt_paid" with the existing data frame
df['total_amt_paid'] = total_amt_paid
df.head()
```

```python
# add new column named "tip_percentange"
tip_percentange = 100 * df['tip']/df['total_bill']
df['tip_percentange'] = tip_percentange
df.head()
```

```python
# roundup the value for "tip_percentange" column with 2 decimel numner
df['tip_percentange'] = np.round(tip_percentange, 2)
df.head()
```

```python
# create another new column named "ratio_of_tips" calculated with the "total_amt_paid" and "tip_percentage"
ratio_of_tips = 100 * df['tip_percentange']/df['total_amt_paid']
df['ratio_of_tips'] = np.round(ratio_of_tips,2)
df.head()
```

```python
# remove the column named "ratio_of_tips" from the data frame
df.drop('ratio_of_tips', 1) 
```

```python
# the best approach to define the final data frame as bellow. This actuall change it parmanently after deleting any column from the data frame
df = df.drop('ratio_of_tips', 1) 
df
```

```python
total_amt_paid = df['tip'] + df['total_bill']
total_amt_paid
```

```python
# add the newly created column named "total_amt_paid" with the existing data frame
df['total_amt_paid'] = total_amt_paid
df.head()
```

```python
# add new column named "tip_percentange"
tip_percentange = 100 * df['tip']/df['total_bill']
df['tip_percentange'] = tip_percentange
df.head()
```

```python
# roundup the value for "tip_percentange" column with 2 decimel numner
df['tip_percentange'] = np.round(tip_percentange, 2)
df.head()
```

```python
# create another new column named "ratio_of_tips" calculated with the "total_amt_paid" and "tip_percentage"
ratio_of_tips = 100 * df['tip_percentange']/df['total_amt_paid']
df['ratio_of_tips'] = np.round(ratio_of_tips,2)
df.head()
```

```python
# remove the column named "ratio_of_tips" from the data frame
df.drop('ratio_of_tips', 1) 
```

```python
# the best approach to define the final data frame as bellow. This actuall change it parmanently after deleting any column from the data frame
df = df.drop('ratio_of_tips', 1) 
df
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Others"
```

```python
quality(16.00, 1.04)
```

```python
df['quality'] = df[['total_bill', 'tip']].apply(lambda df : quality(df['total_bill'], df['tip']), axis=1)
```

```python
df['quality']
```

```python
# Apart from the "lambda expression" we can use "vectorize" method from numpy library where result will same
df['quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
# COmpaire the performances between lambda and vectorize
import timeit
```

```python
# code snippet to be executed only once
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('D:/Download/UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/tips.csv')

def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Others"
'''
```

```python
# code snippet whose execution time is to measured
stmt_one = '''
df['quality'] = df[['total_bill', 'tip']].apply(lambda df : quality(df['total_bill'], df['tip']), axis=1)
'''

stmt_two = '''
df['quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
```

```python
timeit.timeit(setup=setup, stmt=stmt_one, number=1000)
```

```python
timeit.timeit(setup=setup, stmt=stmt_two, number=1000)
```
